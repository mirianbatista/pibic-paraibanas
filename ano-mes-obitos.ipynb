```python
import pandas as pd
```

```python
mulheres_pb = pd.read_csv("paraibanas.csv", sep=";")
mulheres_pb.info()
```

```python
def separaData(data): 
  data = pd.to_datetime(data, format='%Y%m%d')
  data = data.strftime('%Y-%m-%d')
  return data

def separaAno(data): 
  data = data.split('-')[0]
  return data

def separaMes(data): 
  data = data.split('-')[-2]
  return data

def separaDia(data): 
  data = data.split('-')[-1]
  return data

mulheres_pb['DT_SAIDA'] = mulheres_pb.apply(lambda row: separaData(row['DT_SAIDA']), axis=1)
mulheres_pb.head()
```

```python
obitos = mulheres_pb.loc[(mulheres_pb['MORTE']) == 1]
```

```python
obitos['ano_mes'] = pd.to_datetime(obitos['DT_SAIDA']).dt.to_period('M')
obitos.head()
```

```python
ob = obitos.ano_mes.value_counts().to_frame().reset_index().rename(columns = {"index" : "anomes", "ano_mes" : "qtd_anomes"}).sort_values(by='anomes')
ob.to_csv("anomes_ob.csv", index=False)
ob
```
