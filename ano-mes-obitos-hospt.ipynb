```python
import pandas as pd
```

```python
mulheres_pb = pd.read_csv("paraibanas.csv", sep=";")
mulheres_pb.info()
```

```python
def separaData(data): 
  data = pd.to_datetime(data, format='%Y%m%d')
  data = data.strftime('%Y-%m-%d')
  return data

def separaAno(data): 
  data = data.split('-')[0]
  return data

def separaMes(data): 
  data = data.split('-')[-2]
  return data

def separaDia(data): 
  data = data.split('-')[-1]
  return data

mulheres_pb['DT_SAIDA'] = mulheres_pb.apply(lambda row: separaData(row['DT_SAIDA']), axis=1)
mulheres_pb['DT_INTER'] = mulheres_pb.apply(lambda row: separaData(row['DT_INTER']), axis=1)
mulheres_pb.head()
```

```python
obitos = mulheres_pb.loc[(mulheres_pb['MORTE']) == 1]
hospt = mulheres_pb.loc[(mulheres_pb['MORTE']) == 0]
```

```python
obitos['ano_mes_obitos'] = pd.to_datetime(obitos['DT_SAIDA']).dt.to_period('M')
hospt['ano_mes_hospt'] = pd.to_datetime(hospt['DT_INTER']).dt.to_period('M')
```

```python
ob = obitos.ano_mes_obitos.value_counts().to_frame().reset_index().rename(columns = {"index" : "ano_mes_obitos", "ano_mes_obitos" : "qtd_anomes_obitos"}).sort_values(by='ano_mes_obitos')
ob.to_csv("anomes_ob.csv", index=False)
ob
```

```python
hos = hospt.ano_mes_hospt.value_counts().to_frame().reset_index().rename(columns = {"index" : "ano_mes_hospt", "ano_mes_hospt" : "qtd_anomes_hospt"}).sort_values(by='ano_mes_hospt')
hos.to_csv("anomes_hospt.csv", index=False)
hos
```
